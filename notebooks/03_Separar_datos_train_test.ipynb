```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

## Carga de datos limpios

```python
df = pd.read_csv("../data/processed/datos_limpios.csv")
df.head()
```

## Definir X, y, train y test

```python
X = df.drop(["Diagnosis"], axis=1)
y = df["Diagnosis"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=11)

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

## Guardado de datos en 2 .csv diferentes, train y test

```python
# Unimos la parte X y la parte y tanto de train y test para poder guardarlos en 2 dataset diferentes con todas las columnas

train = pd.merge(X_train, y_train, left_index=True, right_index=True)
test = pd.merge(X_test, y_test, left_index=True, right_index=True)
```

```python
# Se guardan en la carpeta data, dentro de otras carpetas, llamadas train y test

train.to_csv("../data/train/train.csv", index=False)
test.to_csv("../data/test/test.csv", index=False)
```
